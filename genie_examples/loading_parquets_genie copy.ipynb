### a simple 'trigger' could be a hit cut of say 5 photon hits in order to accept an event
This ould mean out of 10,000 events simulated the number of events that pass this trigger is:

```python
out_set
```

```python
photons_out_set = out_set['photons']
mc_truth = out_set['mc_truth']
idx = []
max_idx = 0
hit_cut = 5
for i in range(len(photons_out_set)):
    if len(out_set['photons'][i]['t']) >= hit_cut:
        idx.append(i)
        if len(out_set['photons'][i]['t'])  > max_idx:
            max_idx= i

max_hits = len(out_set['photons'][max_idx]['t'])
print(f'The number of events with at least 5 hits is {len(idx)}')
print(f'The event with the most hits is event {max_idx} with {max_hits} hits')
```
