### Showing how to load genie parquets with MC Truth info


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Loading both the primary_set (neutrino info) and prometheus_set (child particle info)

make sure you change paths!

```python

primary_set_path = 'output/genie_events_primary_new_test_simset_99.parquet'
prometheus_set_path = 'output/genie_events_prometheus_new_test_simset_99.parquet'
out_file = 'output/10000events_simset_99.parquet'
prometheus_set = pd.read_parquet(prometheus_set_path)
primary_set = pd.read_parquet(primary_set_path)
out_set = pd.read_parquet(out_file)
```

### lets plot primary info to verify it's in a cylinder:

```python
pos_x = np.array(primary_set['pos_x'])
pos_y = np.array(primary_set['pos_y'])
pos_z = np.array(primary_set['pos_z'])
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(projection='3d')
ax.scatter(pos_x, pos_y, pos_z)
ax.set_title('injection neutrino position in detector coordinates', fontsize=16)
ax.set_xlabel('X (m)')
ax.set_ylabel('Y (m))')
ax.set_zlabel('Z (m))')

plt.show()
```

## cool, and the neutrino directions:

```python
theta = np.array(primary_set['theta'])
phi = np.array(primary_set['phi'])
plt.figure(figsize=(12,8))
plt.hist(phi, bins = 50, edgecolor='black')
plt.title('phi histogram (neutrino direction)')
plt.xlabel('phi', fontsize =16)
plt.ylabel('count', fontsize = 16)
plt.show()
plt.figure(figsize=(12,8))
plt.hist(np.cos(theta), bins = 50, edgecolor='black')
plt.title('cos theta histogram (neutrino direction)')
plt.xlabel('cos theta', fontsize =16)
plt.ylabel('count', fontsize = 16)
plt.show()
```

## cool and the energy:

(specified a $E^{-2}$ flux with gevgen, result may be a $E^-1$ when convolved with cross section?)

```python

energy = np.array(primary_set['e'])
plt.figure(figsize=(12,8))
plt.hist(energy, bins = 50, edgecolor='black')
plt.title('enery (GeV)')
plt.xlabel('energ (GeV)', fontsize =16)
plt.ylabel('count', fontsize = 16)
plt.show()
```

### a simple 'trigger' could be a hit cut of say 5 photon hits in order to accept an event
This ould mean out of 10,000 events simulated the number of events that pass this trigger is:

```python
out_set
```

```python
photons_out_set = out_set['photons']
mc_truth = out_set['mc_truth']
idx = []
max_idx = 0
hit_cut = 5
for i in range(len(photons_out_set)):
    if len(out_set['photons'][i]['t']) >= hit_cut:
        idx.append(i)
        if len(out_set['photons'][i]['t'])  > max_idx:
            max_idx= i

max_hits = len(out_set['photons'][max_idx]['t'])
print(f'The number of events with at least 5 hits is {len(idx)}')
print(f'The event with the most hits is event {max_idx} with {max_hits} hits')
```
